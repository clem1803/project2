```python
import pandas as pd
from sqlalchemy import create_engine
```

## Aircraft data

```python
file_path = "data/WWI/THOR_WWI_AIRCRAFT_GLOSSARY.csv"
data = pd.read_csv(file_path)
```

```python
WWI_aircraft_data = data.drop(["NAME"], axis=1)
```

```python
WWI_aircraft_data = WWI_aircraft_data.rename(columns={"ID": "Aircraft_id", "TYPE": "Aircraft_type"})
```

```python
WWI_aircraft_data.columns
```

```python
WWI_aircraft_data.to_csv("WWI_aircraft_data.csv")
```

## WWI Aerial Attack Data

```python
file_path = "data/WWI/THOR_WWI_CLEAN.csv"
data = pd.read_csv(file_path)
data
```

```python
WWI_aerial_attack_data = data.drop(["OPERATION", "MISSIONNUM", "BDA", "ENEMYACTION", "ROUTEDETAILS",
                                   "ISRCOLLECTED", "WEATHER"], axis=1)
WWI_aerial_attack_data
```

```python
WWI_aerial_attack_data = WWI_aerial_attack_data.drop(["CALLSIGN", "WEAPONSEXPENDED", ], axis=1)
WWI_aerial_attack_data
```

```python
WWI_aerial_attack_data.to_csv("WWI_aerial_attack_data.csv")
```

## WWI Weapon

```python
file_path = "data/WWI/THOR_WWI_WEAPON_GLOSSARY.csv"
data = pd.read_csv(file_path)
data
```

```python
WWI_weapon_data = data.drop(["WEAPON_TYPE", "NBR_BOMBLETS", "ALT_WEAPON_NAME"], axis=1)
WWI_weapon_data
```
