```python
import pandas as pd
from sqlalchemy import create_engine
import psycopg2
```

## Aircraft data

```python
file_path = "raw_data/WWI/THOR_WWI_AIRCRAFT_GLOSSARY.csv"
data = pd.read_csv(file_path)
```

```python
WWI_aircraft_data = data.drop(["NAME"], axis=1)
```

```python
WWI_aircraft_data = WWI_aircraft_data.rename(columns={"ID": "Aircraft_id", "TYPE": "Aircraft_type"})
```

```python
WWI_aircraft_data.columns
```

```python
WWI_aircraft_data.to_csv("WWI_aircraft_data.csv")
```

```python
WWI_aircraft_data
```

## WWI Aerial Attack Data

```python
file_path = "raw_data/WWI/THOR_WWI_CLEAN.csv"
data = pd.read_csv(file_path, encoding="ISO-8859-1")
```

```python
WWI_aerial_attack_data = data.drop(["OPERATION", "MISSIONNUM", "BDA", "ENEMYACTION", "ROUTEDETAILS",
                                   "ISRCOLLECTED", "WEATHER"], axis=1)
```

```python
WWI_aerial_attack_data = WWI_aerial_attack_data.drop(["CALLSIGN", "WEAPONSEXPENDED", ], axis=1)
WWI_aerial_attack_data
```

```python
WWI_aerial_attack_data.to_csv("WWI_aerial_attack_data.csv")
```

```python
WWI_aerial_attack_data.to_json("WWI_aerial_attack_data.json")
```

## WWI Weapon

```python
file_path = "raw_data/WWI/THOR_WWI_WEAPON_GLOSSARY.csv"
data = pd.read_csv(file_path)
```

```python
WWI_weapon_data = data.drop(["WEAPON_TYPE", "NBR_BOMBLETS", "ALT_WEAPON_NAME"], axis=1)
WWI_weapon_data
```

```python
WWI_weapon_data.to_csv("WWI_weapon_data.csv")
```

 # WWII Aerial Attack Data

```python
file_path = "raw_data/WWII/THOR_WWII_DATA_CLEAN.csv"
data = pd.read_csv(file_path, encoding="ISO-8859-1")
```

```python
WWII_aerial_attack_data = data.drop(["MASTER_INDEX_NUMBER", "THEATER", "BDA", "NAF", "TGT_COUNTRY_CODE", "TGT_ID",
                                   "TGT_INDUSTRY_CODE", "TGT_INDUSTRY", "SOURCE_LATITUDE", "SOURCE_LONGITUDE"], axis=1)
```

```python
WWII_aerial_attack_data = WWII_aerial_attack_data.drop(["UNIT_ID", "MDS", "MSN_TYPE", "TGT_PRIORITY",
                                                        "TGT_PRIORITY_EXPLANATION", "AC_ATTACKING", "ALTITUDE", "ALTITUDE_FEET", "NUMBER_OF_HE", "TYPE_OF_HE", "LBS_HE",
                                                        "TONS_OF_HE", "NUMBER_OF_IC", "TYPE_OF_IC", "LBS_IC", "TONS_OF_IC", "NUMBER_OF_FRAG", "TYPE_OF_FRAG", "LBS_FRAG",
                                                        "TONS_OF_FRAG", "TOTAL_LBS", "TOTAL_TONS", "TAKEOFF_BASE","TAKEOFF_COUNTRY", "TAKEOFF_LATITUDE", "TAKEOFF_LONGITUDE",
                                                        "AC_LOST", "AC_DAMAGED", "AC_AIRBORNE", "AC_DROPPING", "TIME_OVER_TARGET", "SIGHTING_METHOD_CODE", "SIGHTING_EXPLANATION",
                                                        "CALLSIGN", "ROUNDS_AMMO", "SPARES_RETURN_AC", "WX_FAIL_AC", "MECH_FAIL_AC", "MISC_FAIL_AC", "TARGET_COMMENT", "MISSION_COMMENTS",
                                                        "SOURCE", "DATABASE_EDIT_COMMENTS"], axis=1)
WWII_aerial_attack_data
```

```python
WWII_aerial_attack_data.to_csv("WWII_aerial_attack_data.csv")
```

```python
WWII_aerial_attack_data.to_json("WWII_aerial_attack_data.json")
```
